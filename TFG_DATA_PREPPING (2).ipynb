# 2. Cleaning the data

```python
df_merged = df_merged[
    df_merged["current_club_name"].notna() &
    ~df_merged["current_club_name"].isin(["Retired", "Unknown"])
]

df_merged.head(10)
```

Eliminate the ID from the player_name column

```python
df_merged["player_name"] = df_merged["player_name"].str.replace(
    r"\s*\(\d+\)$", "", regex=True
)

df_merged.head(5)
```

Eliminate players with no registered market value (irrelevant)

```python
df_merged = df_merged[df_merged["value"].notna() & (df_merged["value"] > 0)]

df_merged.head(5)
```

```python
!pip install rapidfuzz
import pandas as pd
import numpy as np
import re
import unicodedata

# ------------------------------------------------------------
uploaded = files.upload()

for filename in uploaded.keys():
  print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')
  df_uploaded = pd.read_csv(filename)
  print(f'Successfully loaded "{filename}" into a DataFrame.')
  display(df_uploaded.head())
# ------------------------------------------------------------

# ---------------------------
# 1) Name normalization utils
# ---------------------------
def normalize_name(x):
    """Lowercase, strip, remove accents, collapse spaces, remove trailing '(digits)'."""
    if not isinstance(x, str):
        return ""
    x = x.strip()
    x = re.sub(r"\s*\(\d+\)\s*$", "", x)
    x = (
        unicodedata.normalize("NFKD", x)
        .encode("ascii", "ignore")
        .decode("utf-8", errors="ignore")
    )
    x = x.lower().strip()
    x = re.sub(r"\s+", " ", x)
    return x

# ---------------------------------------
# 2) Make working copies + normalize names
# ---------------------------------------
df_perf = df.copy()
df_prof = df_merged.copy()

if "Player" not in df_perf.columns:
    raise KeyError("Performance df (df) must contain a 'Player' column.")
if "player_name" not in df_prof.columns:
    raise KeyError("Profiles df (df_merged) must contain a 'player_name' column.")

df_perf["name_norm"] = df_perf["Player"].map(normalize_name)
df_prof["name_norm"] = df_prof["player_name"].map(normalize_name)

df_perf = df_perf[df_perf["name_norm"] != ""].copy()
df_prof = df_prof[df_prof["name_norm"] != ""].copy()

# -----------------------------------------
# 3) Exact merge first
# -----------------------------------------
df_total = df_perf.merge(
    df_prof,
    on="name_norm",
    how="left",
    suffixes=("", "_profile")
)

profile_match_col = "player_id" if "player_id" in df_total.columns else "player_name"
unmatched_mask = df_total[profile_match_col].isna()

# snapshot exact-match count before fuzzy pass
n_exact_matched = int((~unmatched_mask).sum())

# -----------------------------------------
# 4) Fuzzy matching for the unmatched names
# -----------------------------------------
perf_unmatched_names = df_total.loc[unmatched_mask, "name_norm"].unique().tolist()
prof_names = df_prof["name_norm"].unique().tolist()

try:
    from rapidfuzz import process, fuzz

    def best_fuzzy_match(query, choices):
        m = process.extractOne(query, choices, scorer=fuzz.token_set_ratio)
        return (m[0], float(m[1])) if m else (None, 0.0)

except Exception:
    import difflib

    def best_fuzzy_match(query, choices):
        m = difflib.get_close_matches(query, choices, n=1, cutoff=0)
        if not m:
            return (None, 0.0)
        score = difflib.SequenceMatcher(None, query, m[0]).ratio() * 100.0
        return (m[0], float(score))

THRESHOLD = 90.0

mapping = {}
scores = {}

for q in perf_unmatched_names:
    m, s = best_fuzzy_match(q, prof_names)
    if m is not None and s >= THRESHOLD:
        mapping[q] = m
        scores[q] = s

df_total["name_norm_fuzzy"] = df_total["name_norm"]
df_total.loc[unmatched_mask, "name_norm_fuzzy"] = (
    df_total.loc[unmatched_mask, "name_norm"].map(mapping).fillna(df_total.loc[unmatched_mask, "name_norm"])
)

df_total["fuzzy_matched"] = False
df_total.loc[unmatched_mask, "fuzzy_matched"] = df_total.loc[unmatched_mask, "name_norm"].isin(mapping.keys())

df_total["fuzzy_score"] = np.nan
df_total.loc[unmatched_mask, "fuzzy_score"] = df_total.loc[unmatched_mask, "name_norm"].map(scores)

prof_cols_to_add = [c for c in df_prof.columns if c not in ("name_norm",)]
df_prof_for_merge = df_prof.rename(columns={"name_norm": "name_norm_fuzzy"})

df_total = df_total.merge(
    df_prof_for_merge[["name_norm_fuzzy"] + prof_cols_to_add],
    on="name_norm_fuzzy",
    how="left",
    suffixes=("", "_fuzzy")
)

for col in df_prof.columns:
    if col in ("name_norm",):
        continue
    fuzzy_col = f"{col}_fuzzy"
    if fuzzy_col in df_total.columns and col in df_total.columns:
        df_total[col] = df_total[col].combine_first(df_total[fuzzy_col])

drop_fuzzy_dupes = [c for c in df_total.columns if c.endswith("_fuzzy")]
df_total = df_total.drop(columns=drop_fuzzy_dupes, errors="ignore")

# -----------------------------------------
# 5) MATCH DIAGNOSTICS  ← INSERT FOR THESIS
# -----------------------------------------
total_records   = len(df_total)
n_fuzzy_matched = int(df_total["fuzzy_matched"].sum())
total_matched   = n_exact_matched + n_fuzzy_matched
total_unmatched = int(df_total[profile_match_col].isna().sum())
match_rate      = total_matched / total_records * 100

fuzzy_scores_series = df_total.loc[df_total["fuzzy_matched"], "fuzzy_score"]
avg_fuzzy_score = fuzzy_scores_series.mean()
min_fuzzy_score = fuzzy_scores_series.min()

print("=" * 57)
print("       RAPIDFUZZ MERGE — MATCH DIAGNOSTICS")
print("=" * 57)
print(f"  Total performance records       : {total_records:,}")
print(f"  Exact matches  (FBRef ↔ TM)    : {n_exact_matched:,}")
print(f"  Fuzzy matches  (score ≥ {THRESHOLD:.0f})     : {n_fuzzy_matched:,}")
print(f"  ───────────────────────────────────────────")
print(f"  Total matched                   : {total_matched:,}")
print(f"  Unmatched (no profile found)    : {total_unmatched:,}")
print(f"  Overall match rate              : {match_rate:.1f}%")
print(f"  ───────────────────────────────────────────")
if n_fuzzy_matched > 0:
    print(f"  Avg fuzzy score (matched rows)  : {avg_fuzzy_score:.1f} / 100")
    print(f"  Min fuzzy score accepted        : {min_fuzzy_score:.1f} / 100")
print("=" * 57)

# Sample of unmatched names for manual audit
unmatched_sample = (
    df_total.loc[df_total[profile_match_col].isna(), "Player"]
    .dropna().unique()[:10]

)
if len(unmatched_sample):
    print("\n  Sample unmatched names (first 10):")
    for name in unmatched_sample:
        print(f"    · {name}")

# -----------------------------------------
# 6) Save final dataset
# -----------------------------------------
out_path = "/content/df_total.csv"
df_total.to_csv(out_path, index=False, encoding="utf-8")

out_path
```
